```python
import matplotlib.pyplot as plt
import numpy as np

years = [1990, 2000, 2010]
num_bins = 50

fig, axes = plt.subplots(len(years), 3, figsize=(24, 6 * len(years)))  # 3 columns, each row for one year

global_min = df['IPF'].min()
global_max = df['IPF'].max()
bins = np.linspace(global_min, global_max, num_bins + 1)

for i, year in enumerate(years):
    # All countries
    ax = axes[i, 0]  # Access the correct subplot
    data = df[df['Year'] == year]['IPF']
    counts, _, _ = ax.hist(data, bins=bins, edgecolor='black', cumulative=False, alpha=0.7, label='Frequency')
    cumulative_counts = np.cumsum(counts)
    cumulative_percentage = cumulative_counts / cumulative_counts[-1] * 100

    ax.set_title(f'All Countries - {year}')
    ax.set_xlabel('IPF')
    ax.set_ylabel('Frequency')
    ax.grid(True)

    # Low income
    ax = axes[i, 1]  # Access the correct subplot
    data = df[(df['Year'] == year) & (df['Income_level_2000'] == 'Low Income')]['IPF']
    counts, _, _ = ax.hist(data, bins=bins, edgecolor='black', cumulative=False, alpha=0.7, label='Frequency')
    cumulative_counts = np.cumsum(counts)
    cumulative_percentage = cumulative_counts / cumulative_counts[-1] * 100

    ax.set_title(f'Low Income - {year}')
    ax.set_xlabel('IPF')
    ax.set_ylabel('Frequency')
    ax.grid(True)

    # High income
    ax = axes[i, 2]  # Access the correct subplot
    data = df[(df['Year'] == year) & (df['Income_level_2000'] == 'High Income')]['IPF']
    counts, _, _ = ax.hist(data, bins=bins, edgecolor='black', cumulative=False, alpha=0.7, label='Frequency')
    cumulative_counts = np.cumsum(counts)
    cumulative_percentage = cumulative_counts / cumulative_counts[-1] * 100

    ax.set_title(f'High Income - {year}')
    ax.set_xlabel('IPF')
    ax.set_ylabel('Frequency')
    ax.grid(True)

for ax_row in axes:
    for ax in ax_row:
        ax.set_ylim(0, max(ax.get_ylim()[1] for ax_row in axes for ax in ax_row))  # Normalize y-axis across all plots
        ax.set_xlim(global_min, global_max)

for ax_row in axes:
    for ax in ax_row:
        ax2 = ax.twinx()
        ax2.plot(bins[:-1], cumulative_percentage, color='red', marker='o', linestyle='-', markersize=3)
        ax2.set_ylabel('Cumulative Percentage')
        ax2.axhline(y=80, color='grey', linestyle='--', linewidth=1)
        ax2.set_ylim(0, 100)

fig.suptitle('IPF Distribution by Year and Income Level', fontsize=20, y=1.02)  # Adjust y for spacing, fontsize as needed

fig.tight_layout()
plt.savefig('IPF Distribution by Year and Income Level')
plt.show()
```

# BII calculation

```python
df['IPF_10']=df['IPF_2000']-df['IPF']
df['BII_10']=df['BII_2000']-df['BII']
df['IPF_10_2']=df['IPF_2010']-df['IPF_2000']
df['BII_10_2']=df['BII_2010']-df['BII_2000']
df['IPF_20']=df['IPF_2010']-df['IPF']
df['BII_20']=df['BII_2010']-df['BII']

df['Econ Gain_10']=df['Econ Gain_2000']-df['Econ Gain']
df['Econ Gain_10_2']=df['Econ Gain_2010']-df['Econ Gain_2000']
df['Econ Gain_20']=df['Econ Gain_2010']-df['Econ Gain']

df['Econ Gain_10%']=(df['Econ Gain_10']/df['Econ Gain'])
df['Econ Gain_10_2%']=(df['Econ Gain_10_2']/df['Econ Gain_2000'])
df['Econ Gain_20%']=(df['Econ Gain_20']/df['Econ Gain'])

df.head()
```

```python
def up_down(row,col):
    result=''
    if row[col]>0:
        result='u'
    elif row[col]<0:
        result='d'
    return result

df['IPF_10_ud']=df[df['Year']==1990].apply(lambda row: up_down(row,'IPF_10'),axis=1)
df['BII_10_ud']=df[df['Year']==1990].apply(lambda row: up_down(row,'BII_10'),axis=1)
df['IPF_10_2_ud']=df[df['Year']==1990].apply(lambda row: up_down(row,'IPF_10_2'),axis=1)
df['BII_10_2_ud']=df[df['Year']==1990].apply(lambda row: up_down(row,'BII_10_2'),axis=1)
df['IPF_20_ud']=df[df['Year']==1990].apply(lambda row: up_down(row,'IPF_20'),axis=1)
df['BII_20_ud']=df[df['Year']==1990].apply(lambda row: up_down(row,'BII_20'),axis=1)
df.head()
```
