```python
income=pd.read_excel('/Users/LittleStupidCuteBear/Desktop/Research/20240709 new IPF with rectangular matrix/OGHIST.xls',sheet_name='Country Analytical History')
```

```python
income=income.iloc[10:228,:]
```

```python
columns_list=['Country','Country Name']
for i in range(1987,2019):
    columns_list.append(i)
```

```python
income.columns=columns_list
income
```

```python
df_melted = pd.melt(income, id_vars=['Country','Country Name'], var_name='Year', value_name='Income label')
df_melted
```

```python
df_melted['Income label'] = df_melted['Income label'].apply(lambda x: 'Middle Income' if x in ['LM', 'UM'] else ('Low Income' if x == 'L' else ('High Income' if x == 'H' else x)))
df_melted['Income label'] = df_melted['Income label'].replace('..', '')
df_melted
```

```python
area=pd.read_csv('/Users/LittleStupidCuteBear/Desktop/Research/20240709 new IPF with rectangular matrix/Country_land_area_sq_km.csv')
area.head()
```

```python
result = df_melted.merge(area, left_on=['Country', 'Year'], right_on=['Code', 'Year'], how='left')
result=result[['Country','Year','Income label','Land area (sq. km)']]
df_renamed = result.rename(columns={'Land area (sq. km)': 'Land area'})
df_renamed
```
